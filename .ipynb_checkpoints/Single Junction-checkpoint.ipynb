```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def junction(t0, tf, n, a, i):
    ts = (tf - t0) / n
    tspan = np.arange(t0, tf + ts, ts)  # time span

    phi = np.zeros_like(tspan)
    v = np.zeros_like(tspan)
    # Initial Conditions
    phi[0] = 0
    v[0] = 0

    for j in range(1, n + 1):
        phi[j] = phi[j - 1] + v[j - 1] * ts
        v[j] = v[j - 1] + (i - np.sin(phi[j - 1]) - a * v[j - 1]) * ts
    return phi, v, tspan
```

### Josephson Junction Equation

$\ddot{\phi}+\Gamma\dot{\phi}+sin{(\phi)}=i$


```python
a = 0.5
i = 1.1
phi, v, tspan = junction(0, 200, 20000, a, i)

phase = np.sin(phi)

plt.figure(1)
plt.title(r'$\phi$(t) $\alpha$ = {} & $I_0$ = {}'.format(a, i))
plt.plot(tspan, phase)

plt.figure(2)
plt.title(r'V(t) $\alpha$ = {} & $I_0$ = {}'.format(a, i))
plt.plot(tspan, v)

plt.show()
```
